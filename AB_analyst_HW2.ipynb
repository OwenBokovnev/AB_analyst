Гипотезы для ozon.ru:

1. Если увеличить процентную ставку по некоторым категориям товаров, то, несмотря на некоторый отток продавцов, прибыль увеличится на 10-15%
2. Если снизить процентную ставку по некоторым категориям товаров, то приток новых продавцов составит 5-10%
3. Если убрать процентную ставку по некоторым категориям товаров для определенной категории продавцов (люди с ограниченными возможностями, мелкий бизнес и т.п.), то приток новых продавцов составит 10-15%
4. Если ввести льготные процентные ставки по товарам для некоторых регионов страны, то приток новых продавцов увеличится на 5-7%
5. Если поднять процентные ставки по товарам для некоторых регионов страны, то, несмотря на некоторый отток продавцов, прибыль увеличится на 5-7%
6. Если снизить стоимость франшизы на 10%, то прибыль увеличится на 5-7%
7. Если увеличить стоимость франшизы на 5%, то прибыль увеличится на 7-9%
